# 1) 다양한 단어의 표현 방법

### 1. 단어의 표현 방법

단어의 표현 방법을 두 가지로 분류하면 다음과 같다.

* 국소 표현(Local Representation) = 이산 표현(Discrete Representation)
* 분산 표현(Distributed Representation  = 연속 표현(Continuous Representation)

#### 국소 표현(Local Representation) = 이산 표현(Discrete Representation)

* 국소 표현은 해당 단어 자체만 보고 특정값을 매핑하여 단어를 표현하는 방법

Ex) puppy, cute, lovely 라는 단어가 있을 때 각각 단어에 1번, 2번, 3번 과 같이 숫자를 부여하는 방식

#### 분산 표현(Distributed Representation  = 연속 표현(Continuous Representation)

* 국소 표현과 달리 해당 단어와 같이 자주 등장하는 단어들을 고려하여 표현

Ex) Puppy, Cute, Lovely 라는 단어가 있을 때, Puppy라는 단어가 주로 Cute, Lovely 와 같이 나온다고 가정했을 때\
Puppy라는 단어는 cute, lovely한 느낌이다라는 것을 단어로 정의

#### 국소 표현(Local Representation)  vs 분산 표현(Distributed Representaion)

* 국소 표현 : 단어의 의미, 뉘앙스를 고려하지 않고 표현
* 분산 표현 : 단어의 의미, 뉘앙스를 고려하여 표현
