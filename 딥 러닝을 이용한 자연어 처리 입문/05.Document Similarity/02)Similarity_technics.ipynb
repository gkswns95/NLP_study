### 1. 유클리드 거리(Euclidean distance)

코사인 유사도, 자카드 유사도만큼 유용한 방법은 아니지만 유클리드 거리라는 문서간 유사도를 구하는 방법이

있다는 것을 개념적으로 알아두면 좋다.

두 개의 점 p 와 q 둘 사이의 거리를 계산하는 유클리드 거리 공식은 아래와 같다.

Ex) 2차원 공간이라고 생각하면 아래처럼 p와q의 거리를 계산할 수 있다.

이제 유클리드 유사도 방식을 사용하여 여러 문서에 대해서 특정 문서와의 유사도를 구하는 실습을 진행하겠다.

* 문서1,2,3 에 대한 DTM

* 문서 Q(비교 문서)

문서 Q에 대해서 나머지 문서들과의 유클리드 유사도를 구하는 실습

```python
import numpy as np
def dist(x,y):
    return np.sqrt(np.sum((x-y)**2))

doc_1 = np.array((2,3,0,1))
doc_2 = np.array((1,2,3,1))
doc_3 = np.array((2,1,2,2))
doc_Q = np.array((1,1,0,1))

print(' 문서1과 문서Q에 대한 유클리드 유사도 :',dist(doc_1,doc_Q))
print(' 문서2과 문서Q에 대한 유클리드 유사도 :',dist(doc_2,doc_Q))
print(' 문서3과 문서Q에 대한 유클리드 유사도 :',dist(doc_3,doc_Q))
```

### 2. 자카드 유사도(Jaccard similarity)

자카드 유사도는 문서를 각각 집합으로 보고 두개의 집합이 가지는 교집합, 즉 공통으로 가지고 있는 원소를 이용하여

두 집합의 유사도를 구하고자 하는 방식이다.

<자카드 유사도 실습코드>

```python
#두 문서 모두에서 등장한 단어 apple, banana 총 2개
doc1 ="apple banana everyone like likey watch card holder"
doc2 ="apple banana coupon passport love you"

#토큰화 작업
tokenized_doc1 = doc1.split()
tokenized_doc2 = doc2.split()

#토큰화 결과 확인
print(tokenized_doc1)
print(tokenized_doc2)
```

* 문서1과 문서2의 합집합

```python
union = set(tokenized_doc1).union(set(tokenized_doc2))
print(union)
```

* 문서1과 문서2의 교집합

```python
intersection = set(tokenized_doc1).intersection(set(tokenized_doc2))
print(intersection)
```

```python
print(len(intersection)/len(union)) # 2를 12로 나눔
```
